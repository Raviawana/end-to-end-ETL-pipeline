```python
# =====================================
# WIDGETS
# =====================================
dbutils.widgets.text(
    "config_path",
    "/Workspace/Users/ud3041@gmail.com/end-to-end-ETL-pipeline/databricks/gold/config.json",
    "Config File Path"
)

config_path = dbutils.widgets.get("config_path")

# =====================================
# LOAD CONFIG
# =====================================
import json

config_path = dbutils.widgets.get("config_path")
with open(config_path, "r") as f:
    config = json.load(f)

CATALOG = config["catalog"]
SILVER_SCHEMA = config["silver_schema"]
GOLD_SCHEMA = config["gold_schema"]
TABLES = config["tables"]


# =====================================
# SILVER → GOLD COPY
# =====================================
for table in TABLES:
    print(f"Promoting table: {table}")

    silver_df = spark.table(f"`{CATALOG}`.{SILVER_SCHEMA}.{table}")

    silver_df.write \
        .format("delta") \
        .mode("overwrite") \
        .option("overwriteSchema", "true") \
        .saveAsTable(f"`{CATALOG}`.{GOLD_SCHEMA}.{table}")

print("Silver → Gold promotion completed successfully")
```
